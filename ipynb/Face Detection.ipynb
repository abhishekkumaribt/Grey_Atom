```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = r"C:\Users\eakhumb\Documents\Python\Images\Face Detection\img1.jpg"
```

```python
image = cv2.imread(img_path)
```

```python
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
facecascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = facecascade.detectMultiScale(grey)
```

```python
disimage = image.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(disimage, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(disimage)
```

```python
image = cv2.imread(r"C:\Users\eakhumb\Documents\Python\Images\Face Detection\face3.jpg")
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
eyecascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
eyeglassescascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye_tree_eyeglasses.xml")
eyes = eyecascade.detectMultiScale(grey, scaleFactor=1.3, minNeighbors=1, minSize=(60,60))
```

```python
disimage = image.copy()
for (x, y, w, h) in eyes:
    cv2.rectangle(disimage, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(disimage)
```
